```python
import numpy as np
import matplotlib.pyplot as plt
```

#We have wave function 
$\psi$(x)=$\sqrt\frac{2}{a}\sin(\frac{n\pi x}{a})$


```python
def psi(lim1,lim2,width_of_well,n):#lim1,lim2:boundary
    x=np.linspace(lim1,lim2,1000)
    a=width_of_well
    l=(2/a)**1/2
    y=l*np.sin(n*np.pi*x/a)
    plt.plot(x,y)
    plt.grid()
```

```python
psi(0,2*np.pi,2*np.pi,1)#for n=1
psi(0,2*np.pi,2*np.pi,2)#for n=2
psi(0,2*np.pi,2*np.pi,3)#for n=3
```
