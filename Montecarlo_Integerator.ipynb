```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
def montecarlo_integerator(func,nsample,a,b):
    x=np.linspace(.8,3,100)
    y=func(x)
    ymin,ymax=y.min(),y.max()
    xrvs=ss.uniform.rvs(a,b-a,nsample)
    yrvs=ss.uniform.rvs(ymin,ymax,nsample)
    xs=[]
    ys=[]
    for i in range(nsample):
        if yrvs[i]<func(xrvs[i]):
            xs.append(xrvs[i])
            ys.append(yrvs[i])
    plt.plot(xrvs,yrvs,',',color='C0',label='all point')
    plt.plot(xs,ys,',',color='C1',label='under curve')
    plt.plot(x,y,'k',label='func')
    plt.legend(loc='lower right')
    return ymax*(b-a)*len(ys)/nsample
```

```python
#define your own function to integerate
def func(x):
    return 1/(1+np.sinh(2*x)*np.log(x)**2)
```

```python
montecarlo_integerator(func,25000,.8,3)
```

```python
def func2(x):
    return np.exp(x)
```

```python
montecarlo_integerator(func2,25000,.8,3)
```
