```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Trapazoidal_integerator(func,a,b):#a,b:limit of integeration
    x=np.linspace(a,b,100)
    area=0
    for i in range(len(x)-1):
        b=x[i]
        c=x[i+1]
        s=(c-b)*(func(c)+func(b))/2
        area+=s
    plt.figure(figsize=(13,8))
    plt.plot(x,func(x))
    plt.stem(x, func(x), use_line_collection=True)
    plt.title('Trapazoidal rule Visualization')
    return area
```

```python
#define your own function for integeration
def func(x):
    return 1/(1+np.sinh(2*x)*np.log(x)**2)
```

```python
Trapazoidal_integerator(func,.8,3)
```
